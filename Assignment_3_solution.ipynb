###  1Q.Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce() 

```python
# Reduce will produce a single result
def my_reduce(func, my_lst):
    res = my_lst[0]
    for items in my_lst[1:]:
        res = func(res, items)
    return res
```

```python
#testing myreduce()
def sum(x,y): 
    return x + y
```

```python
print (my_reduce(sum,[1, 2, 3, 4]))
```

### 1.2Q Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```python
def my_filter(my_func, sequence):
    result = []# iterate over sequence of items in sequence and apply filter function
    for val in sequence:
        if my_func(val):
            result.append(val)
    return result
```

```python
# test myfilter function
def ispositive(x):
    if (x <= 0): 
        return False 
    else: 
        return True
```

```python
print(my_filter(ispositive,[1,-1,3,4,-5]))
```
